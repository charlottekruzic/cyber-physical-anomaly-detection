### Colonne label et label_n

```python
for i in range(len(df_all)):
    print(f"\n{df_nom[i]} : {df_all[i][['label', 'label_n']].value_counts()}")
```

On peut voir que label est à 0 lors d'une situation normale, et à 1 lors des attaques.  
C'est ce qu'il fallait, pas besoin de nettoyer cela.

Le fichier normal ne contient pas de label = 1, contrairement aux fichiers d'attaque.  
Le fichier est donc bien un fichier représentant le fonctionnement normale des échanges réseau.

Les différentes valeurs de label :  
- normal : pas d'attaque
- MITM (= Man-in-the-Middle) : L'attaquant intercepte et peut modifier les communications entre deux appareils.
- physical fault : Anomalies physiques non causée par une cyberattaque \
--> Normalement non détectables dans les données réseau
- anomaly : autre anomalies non MITM et pas un défaut physique.
- DoS (=Denial of service) : L'attaquant inonde le service de requêtes/paquets pour le surcharger et le rendre indisponible.
- scan : L'attaquant cherche à identifier les ports ouverts et les vulnérabilités potentielles avant une attaque.

Les colonnes sport, dport et flags sont actuellement représenteés comme des colonnes avec des valeurs continues.  
Cependant, cela n'a pas de sens.

```python
# nombre de valeurs unique par colonne pour chaque dataset
columns_of_interest = ['dport', 'sport', 'flags']
for i in range(len(df_all)):
    print(f"{df_nom[i]} :")
    for col in columns_of_interest:
        print(f"- '{col}' : {df_all[i][col].nunique()} valeurs uniques")
    print()
```

Les colonnes sport et dport ont entre 5000 et 10000 valeurs possibles différentes, il n'est donc pas possible de faire un One-Hot.  
Nous allons donc regrouper les IP par catégories : "System Ports (0-1023), User Ports (1024-49151), et the Dynamic and/or Private Ports (49152-65535)"  
Source : https://www.iana.org/assignments/service-names-port-numbers/service-names-port-numbers.xhtml

Pour la colonne flags, comme il y a moins de 10 valeurs possibles, nous les transformons en catégories, et nous ferons peut etre un one-hot plus tard avant les modèles.

## Nettoyage (3)

### Discrétisation des ports

```python
def get_categorie_ports(port):
    if pd.isna(port):
        return "inconnu" # remplacement des valeurs manquantes
    if port < 1024:
        return "system_ports"
    elif port <= 49151:
        return "user_ports"
    else:
        return "dynamic/private_ports"
```

```python
for i in range(len(df_all)):
    for col in ['sport', 'dport']:
        df_all[i][col] = df_all[i][col].apply(get_categorie_ports).astype('category')
```

Les colonnes catégoriques ip_s, ip_d, modbus_fn, et modbus_response ont des valeurs manquantes.  
Nous allons remplacer ces valeurs manquantes par "Inconnue" pour ne pas perdre d'informations.

Pour les colonnes numériques n_pkt_src, et n_pkt_dst ayant des valeurs manquantes, nous allons remplacer par la médiane.


## Nettoyage (4)

```python
for df in df_all:
    for col in ['ip_s', 'ip_d', 'modbus_fn', 'modbus_response']:
        df[col] = df[col].fillna("inconnue")
    for col in ['n_pkt_src', 'n_pkt_dst']: # La colonne redevient de type float
        df[col] = df[col].median()
```
